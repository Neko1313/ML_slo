```python
import numpy as np
import pandas as pd
dict_data = {}
def read_file(path:str):
    for i in range(6,14):
        data = np.load(path+f"{i}.npy", allow_pickle=True)
        for j in data.tolist().keys():
            dict_data[j] = data.tolist()[j]
```

```python
read_file("./data_2/")
```

```python
indices = []
suffix_to_find = "_Name"
for i, word in enumerate(list(dict_data.keys())):
    if word.endswith(suffix_to_find):
        indices.append(int(word.split("_")[0]))
```

```python
indices.sort()
```

```python
keys_dict = [[f"{i}_Name",f"{i}_A",f"{i}_B",f"{i}_C",f"{i}_SUM",f"{i}_time",f"{i}_KBO",f"{i}_KBZ",f"{i}_open|close: 0|1"] for i in indices]
```

```python
list_dataframe =[]
```

```python
for Name, A, B, C, SUM, time, KBO, KBZ, open_cloce in keys_dict:
    dict_prom = {}
    dict_prom["A"] = dict_data[A]
    dict_prom["B"] = dict_data[B]
    dict_prom["C"] = dict_data[C]
    dict_prom["SUM"] = dict_data[SUM]
    dict_prom["time"] = dict_data[time]
    dict_prom["KBO"] = dict_data[KBO]
    dict_prom["KBZ"] = dict_data[KBZ]
    dict_prom["open_cloce"] = dict_data[open_cloce]
    df_prom = pd.DataFrame(dict_prom)
    list_dataframe.append(df_prom)
```

```python
list_dataframe[3] 
```

```python
list_dataframe_name = [f"{i}" for i in range(len(list_dataframe))]
```

```python
result = pd.concat(list_dataframe, keys=list_dataframe_name)
```
